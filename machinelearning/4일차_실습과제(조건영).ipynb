```python
# 모듈 로드
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_digits
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 데이터 불러오기
digits = load_digits()
```

```python
# 데이터 타겟 나누기
data=digits['data']
target=digits['target']
```

```python
# 데이터를 입력하면 예측값과 숫자 그래프 출력
def print_newdata(a):
    print(f"예측 숫자는 {model.predict(a)}")
    plt.gray()
    plt.matshow([a[0][0:7],a[0][8:15],a[0][16:23],a[0][24:31],a[0][32:39],a[0][40:47],a[0][48:55],a[0][56:63]])
    plt.show()
```

```python
# 숫자 그림 출력 함수
def print_image(n):
    print(f"해당 {n}의 target은 {target[n]}")
    plt.gray()
    plt.matshow(digits.images[n])
    plt.show()
```

```python
# 훈련/테스트 데이터 나누기
train_data,test_data,train_target,test_target=train_test_split(data,target,random_state=42)
```

```python
# 스케일링
ss=StandardScaler().fit(train_data)
train_scaled=ss.transform(train_data)
test_scaled=ss.transform(test_data)
```

# SVC

```python
# 모델 객체 생성 및 훈련
model=SVC(C=1,gamma='scale').fit(train_data, train_target)
```

```python
# 모델 평가
print(f"train의 정확도: {model.score(train_data, train_target)}")
print(f"test의 정확도: {model.score(test_data, test_target)}")
```

```python
# 모델 예측
model.predict(test_data)
```

```python
# 원본 데이터를 입력하여 예측해보기
model.predict([data[1769]])
```

```python
# 예측한 값이 맞는지 그림을 통해 확인
print_image(1769)
```

```python
# 최적 훈련 횟수 알아보기
train_acc=[]
test_acc=[]
for i in range(0,600):
    model=LogisticRegression(max_iter=i).fit(train_scaled,train_target)
    train_acc.append(model.score(train_scaled,train_target))
    test_acc.append(model.score(test_scaled,test_target))
```

```python
# 그래프 그리기
plt.figure(figsize=(15,5))
plt.plot(train_acc)
plt.plot(test_acc)
plt.xlabel('max_iter')
plt.ylabel('accuaracy')
plt.show()
```
