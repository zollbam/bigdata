## Scikit-learn Dataset 사용
- 반환값
  - dict와 유사
  

- 종류
  - load_dataset() : sikit_learn 저장소에서 로딩
  - fetch_XXX() : dataset 저장소에서 가져오기

```python
# 모듈 로딩
import pandas as pd
from sklearn.datasets import load_diabetes
```

```python
# 당뇨병 유무 => 총 8개의 특성 + 1개의 라벨
data=load_diabetes()
type(data)
```

- Bunch라는 타입으로 출력됨

```python
# dict에서 key로 value접근
data['DESCR']
```

```python
# 특성만 추출
print(f"배열\n{data.data}")
print(f"데프\n{pd.DataFrame(data.data)}")
```

```python
# 타겟만 추출
print(f"배열\n{data.target}")
print(f"데프\n{pd.DataFrame(data.target)}")
```

- load_xxx()들은 딕셔너리 형태인듯

```python
# get()함수를 이용한 값 추출
print(data.get('data'))
print(data.get('target'))
```
