```python
# 모듈로딩
import pandas as pd
import numpy as np
```

```python
# 파일 관련 변수
DIR='../Data/'
FILE_NAME='titanic.csv'
```

```python
# 파일 불러오기
titanic=pd.read_csv(DIR+FILE_NAME)
titanic

# PassengerId => 승객아이디(int64) => 계산할 필요도 카테고리로 묶을 필요가 없어보여 object로
# Survived => 생존여부(int64) 0은 구조X 1은 구조O => 범주형으로 바꾸면 좋을거 같아 category로
# Pclass => 배자리 등급(int64) => 범주형으로 바꾸면 좋을거 같아 category로
# Name => 승객이름(object) => 그대로 유지
# Sex => 성별(object) => object보다는 범주형이 좋을 듯하여 category로
# Age => 나이(float64) => 20대 30대 이렇게 해서 범주형으로 해보면 분석시 좋을 듯하여 category로
# SibSp => 타이타닉호에 탑승한 형제자매/배우자(int64) => category로
# Parch => 타이타닉호에 탑승한 부모/자녀의 수(int64) => category로
# Ticket => 티켓숫자(object) => 그대로 유지
# Fare => 요금(float64) => 나중에 계산할때 유용하게 쓰일 실수형 데이터이므로 유지
# Cabin => 방숫자(object) => 그대로 유지
# Embarked => 승선항(object) C(Cherbourg), Q(Queenstown), S(Southampton) => category로
```

```python
# 정보확인
titanic.info()
```

### Age열을 연령대로 바꾸고 Age열 삭제 후 Age_gro열 만들기

```python
# 연령대 확인을 위한 작업
titanic['Age'].value_counts().iloc[10:15]

# 0~70대까지 있는 것을 확인하고 NaN값도 있음
```

```python
# 연령대 열 생성을 위한 작업
age_gr=[]
for a in titanic['Age']:
    if pd.isna(a):
        age_gr.append('X')
    else:
        age_gr.append(str(a)[0]+'0')
titanic['Age']=age_gr
titanic.rename(columns={'Age':'Age_gro'},inplace=True)
titanic

# Age열의 값이 na가 아니면 제일 앞 숫자만 추출하여 0을 더하면 00 10 40 등 연령대가 됨
# NaN값은 나이 미상의 의미로 X라고 설정함
# for문이 완료되면 연령대 리스트가 생성되고 이 리스트를 바로 Age열 값과 변경함
# 변경후 열 이름을  Age_gro로 바꿈
```

```python
# Age_gro열의 고유값 확인
titanic.Age_gro.unique()

# 내가 원하는 대로 바뀐 것을 확인
```

### 타입 변경 하기
- PassengerId -> int64 => object
- Survived -> int64 => category
- Pclass -> int64 => category
- Name -> object => object
- Sex -> object => category
- Age_gro -> float64 => category
- SibSp -> int64 => category
- Parch -> int64 => category
- Ticket -> object => object
- Fare -> float64 => float64
- Cabin -> object => object
- Embarked -> object => category

```python
# 타입변경 함수 만들어보기
def chanty(df):
    as_ty=['object','category','category','object','category','category','category','category','object','float64','object','category']
    for idx in range(len(df.columns)):
        col=df.columns[idx]
        df[col]=df[col].astype(as_ty[idx])
```

```python
# 함수 사용으로 열 타입 변경
chanty(titanic)
```

```python
# 정보 확인
titanic.info()

# 원하는 타입으로 변경 완료
```

### 분석
- 구조여부별 연령대별 성별에 대한 분석
    - 구조된/비구조된 사람중에 연령대별 확률을 구하고 싶음

```python
# 구조여부별 연령대별 성별의 인원 수
surv_su=pd.DataFrame(titanic.groupby(['Survived','Age_gro','Sex']).count()['PassengerId'])
surv_su
```

```python
surv_su.reset_index()

# 이렇게 인덱스를 빼내어서 계산도 가능해 보임!!!!
```

```python
# surv_su에서 Survived가 0인 데이터만 추출하여 확률 구하기
surv0_su=surv_su['PassengerId'][:22]
surv0_pro=round(surv0_su/sum(surv_su['PassengerId'][:22])*100,2)
surv0_pro
```

```python
# surv_su에서 Survived가 1인 데이터만 추출하여 확률 구하기
surv1_su=surv_su['PassengerId'][22:]
surv1_pro=round(surv1_su/sum(surv_su['PassengerId'][22:])*100,2)
surv1_pro
```

```python
# surv0_pro와 surv1_pro 다시 합치기
surv_pro=pd.concat([surv0_pro,surv1_pro])
surv_pro
```

```python
# 열 결합으로 개수와 비율 한눈에 보기
surv_su_pro=pd.concat([surv_su,surv_pro],axis=1)
surv_su_pro

# 즉, 구조 되지 못한(죽음) 사람중에 20대 남자가 22.95%에 달함
# 구조된 사람중에 20대 여자가 15.79%
```

### 분석
- 연령대별 성별 구조여부
    - 연령대별에서 성별에 따른 구조여부 확률로 분석하고 싶음

```python
# 연령대별 성별 구조여부 인원 수
a_s_s_su=titanic.groupby(['Age_gro','Sex','Survived']).count()['PassengerId']
a_s_s_su
```

```python
# 연령별 성별 인원수
a_s_su=titanic.groupby(['Age_gro','Sex']).count()['PassengerId']
a_s_su
```

```python
# 연령별 성별 구조요부의 확률 구하기
a_s_s_pro=(a_s_s_su.div(a_s_su)*100).round(2)
a_s_s_pro

# 0대 아이들은 100% 구조된것을 확인
# 10대 남성들은 83.87%가 구조되지 못함
# 연령대별 남성보다 여성의 구조비율이 높음
```

```python
# 열결합
a_s_s_su_pro=pd.concat([a_s_s_su,a_s_s_pro],axis=1)
a_s_s_su_pro
```
